```python
#importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
#reading companies data
companies = pd.read_csv('companies.txt',sep="\t",encoding="ISO-8859-1")
companies
```

```python
#reading rounds2 data
rounds2=pd.read_csv("rounds2.csv",encoding="ISO-8859-1")
rounds2
```

```python
#creating a master_frame using append
master_frame=pd.concat([rounds2,companies],axis=1)
master_frame
```

```python
#encoding to utf-8 which is codec for all languages and decode to english
master_frame['company_permalink']=master_frame.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
master_frame
```

```python
###### as now we will be working master_frame so cleaning the data#####
```

```python
#Checking for NAN values
master_frame.isnull().sum(axis=0)
```

```python
#calculating the NaN %age
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame.shape
```

```python
#removing rows
master_frame=master_frame[master_frame.isnull().sum(axis=1) <= 5]
```

```python
#Calculating the %age after removing rows
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#removing columns
master_frame=master_frame.drop(['funding_round_code','status','state_code','founded_at','funded_at','region','homepage_url','city'],axis=1)
```

```python
#percentage after removing columns
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#removing the NaN from raised_amount_usd
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
```

```python
#dropping NaN values from country code column
master_frame=master_frame.dropna(subset=['country_code'])
```

```python
#dropping NaN values from category_list column
master_frame=master_frame.dropna(subset=['category_list'])
```

```python
#country_code and category_list cant be dropped and manipulated as these two columns are necessary for our analysis
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#master_frame after cleansing the data
master_frame
```

```python
#using groupby to see the funding round type
Funding_by_segment=master_frame.groupby('funding_round_type')['raised_amount_usd'].mean().astype(int)
Funding_by_segment
```

```python
#top9 countries based on the total investment amount each country received
top9=master_frame.loc[master_frame['funding_round_type']=='venture'].groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False).reset_index()
top9.head(9)
```

```python
#splitting the category column in order to create our primary sector as per the requirement
master_frame['primary_sector'],master_frame['category_list1']=master_frame['category_list'].str.split('|',1).str
master_frame=master_frame.drop('category_list1',axis=1)
```

```python
#Primary sector is present in the master_frame
master_frame
```

```python
#reading the mapping file
mapping=pd.read_csv("mapping.csv",encoding="ISO-8859-1")
mapping
```

```python
#transforming the data in order to map it out against the primary sector to get the main sector
mapping=pd.melt(mapping,id_vars=['category_list'],var_name='main_sector',value_name='value')
mapping=mapping[mapping['value']==1].drop('value',axis=1)
mapping
```

```python
# merging with the master_frame
master_frame = (pd.merge(master_frame, mapping,  how='inner', left_on=['primary_sector'], right_on = ['category_list']))
master_frame
```
